```python
def divide_data(emg,restimulus,experiment_number):
    
    
    
    index_for_chossen_restimulus_number = np.where(restimulus == experiment_number)
    index_for_chossen_restimulus_number = np.array (index_for_chossen_restimulus_number)
    
    index_for_number_before_restimulus_number = np.where(restimulus == (experiment_number -1))
    last_index = index_for_number_before_restimulus_number[0][-1]
    
    
    #after subtraction:
    after_subtraction = np.empty(index_for_chossen_restimulus_number.shape[1])
    after_subtraction[0]=  index_for_chossen_restimulus_number[0][0]
    for i in range(index_for_chossen_restimulus_number.shape[1]-1):
        after_subtraction[i+1] = index_for_chossen_restimulus_number[0][i+1]-index_for_chossen_restimulus_number[0][i]
    
    #get the actual index for the beginning for each trail:
    index = np.where(after_subtraction > 1)
    index = np.array(index)
    
    #get the beginning point for each movement:
    beginning_point = np.empty(index.shape[1])
    for i in range(index.shape[1]):
        beginning_point[i] = index_for_chossen_restimulus_number[0][index[0][i]]
        
    
    #get the end point:
    end_index= np.empty(index.shape[1])
    for i in range(index.shape[1]-1):
        end_index[i] = int(index[0][i+1]-1)
    end_index[-1] = after_subtraction.shape[0]-1

    end_point = np.empty(index.shape[1])
    for i in range(index.shape[1]):
        end_point[i]= index_for_chossen_restimulus_number[0][int(end_index[i])]
    
    
    if experiment_number == 1:
        #get the training and testing sets:
        first_trail_res = restimulus[0:int(end_point[0]+1)]
        first_trail_emg = emg[0:int(end_point[0]+1),:] 
    
        second_trail_res = restimulus[int(end_point[0]+1):int(end_point[1]+1)]
        second_trail_emg = emg[int(end_point[0]+1):int(end_point[1]+1),:]

        third_trail_res = restimulus[int(end_point[1]+1):int(end_point[2]+1)]
        third_trail_emg = emg[int(end_point[1]+1):int(end_point[2]+1),:]
    
        fourth_trail_res = restimulus[int(end_point[2]+1):int(end_point[3]+1)]
        fourth_trail_emg = emg[int(end_point[2]+1):int(end_point[3]+1),:]

        fifth_trail_res = restimulus[int(end_point[3]+1):int(end_point[4]+1)]
        fifth_trail_emg = emg[int(end_point[3]+1):int(end_point[4]+1),:]
    
        sixth_trail_res = restimulus[int(end_point[4]+1):int(end_point[5]+1)]
        sixth_trail_emg = emg[int(end_point[4]+1):int(end_point[5]+1),:]
    
    else:
        #get the training and testing sets:
        first_trail_res = restimulus[last_index+1:int(end_point[0]+1)]
        first_trail_emg = emg[last_index+1:int(end_point[0]+1),:] 
    
        second_trail_res = restimulus[int(end_point[0]+1):int(end_point[1]+1)]
        second_trail_emg = emg[int(end_point[0]+1):int(end_point[1]+1),:]

        third_trail_res = restimulus[int(end_point[1]+1):int(end_point[2]+1)]
        third_trail_emg = emg[int(end_point[1]+1):int(end_point[2]+1),:]
    
        fourth_trail_res = restimulus[int(end_point[2]+1):int(end_point[3]+1)]
        fourth_trail_emg = emg[int(end_point[2]+1):int(end_point[3]+1),:]

        fifth_trail_res = restimulus[int(end_point[3]+1):int(end_point[4]+1)]
        fifth_trail_emg = emg[int(end_point[3]+1):int(end_point[4]+1),:]
    
        sixth_trail_res = restimulus[int(end_point[4]+1):int(end_point[5]+1)]
        sixth_trail_emg = emg[int(end_point[4]+1):int(end_point[5]+1),:]
  
    emg_training_ds = np.concatenate((first_trail_emg,third_trail_emg,fourth_trail_emg,sixth_trail_emg),axis=0)
    res_training_ds = np.concatenate((first_trail_res,third_trail_res,fourth_trail_res,sixth_trail_res),axis=0)
    emg_testing_ds = np.concatenate((second_trail_emg,fifth_trail_emg),axis=0)
    res_testing_ds = np.concatenate((second_trail_res,fifth_trail_res),axis=0)

    return (emg_training_ds,res_training_ds, emg_testing_ds,res_testing_ds)
```
